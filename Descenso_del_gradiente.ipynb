## Descenso del gradiente

## Dependencias

```
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

### Gr치fica en 3D de nuestra funci칩n de coste

```
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x,y):
  return x**2 + y**2;

res = 100

X = np.linspace(-5, 5, res)
Y = np.linspace(-5, 5, res)

X, Y = np.meshgrid(X, Y)

Z = f(X,Y) 

# Gr치ficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap=cm.cool,
                       linewidth=0, antialiased=False)
# a침ade la barra lateral para identificar veisualmente los altos y bajos de la grafica
fig.colorbar(surf)

```

## Descenso del gradiente
### Se acerca al minimo local mas cercano, es muy util cuando solo hay un solo minimo local

```
level_map = np.linspace(np.min(Z), np.max(Z),res) 
plt.contourf(X, Y, Z, levels=level_map,cmap=cm.cool)
plt.colorbar()
plt.title('Descenso del gradiente')

def derivate(_p,p):
  return  (f(_p[0],_p[1]) - f(p[0],p[1])) / h

# Definiendo elos puntos aletarios
p = np.random.rand(2) * 8 - 4 # generar dos valores aleatorios
print("p > ", p)

# Graficando en los puntos aletarios generados, ubicandolo el punto en el mapa
plt.plot(p[0],p[1],'o', c='k')

# Lerning rate, valor constante que alterara el gradiante
lr = 0.01
# Incremento constante
h = 0.01

grad = np.zeros(2)

for i in range(10000):
  for idx, val in enumerate(p): 
    _p = np.copy(p)

    _p[idx] = _p[idx] + h;

    dp = derivate(_p,p) 

    grad[idx] = dp

  p = p - lr * grad

  if(i % 10 == 0):
    plt.plot(p[0],p[1],'o', c='r')

plt.plot(p[0],p[1],'o', c='w')
plt.show()

print("El punto m칤nimo se encuentra en: ", p)
```
